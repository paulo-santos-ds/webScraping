```python
%%time
base = {'liq diaria':[],
       'ultimo rend':[],
       'yield':[],
       'patrim liquido':[],
       'valor patrim':[],
       'rent mes':[],
       'p/vp': []}
index = list(base.keys())
count = 0

for fii in nome_fii:
    html = requests.get("https://www.fundsexplorer.com.br/funds/"+fii.lower()).content 
    soup = BeautifulSoup(html,'html.parser')
    texto = soup.prettify()
    texto = texto.split('<div class="carousel-cell">')
    
    for i in range(1,7):
        aux = texto[i].split('<span class="indicator-value">\n          ')[-1]
        base[index[i-1]].append(aux.split('\n')[0])
        
    aux = texto[7].split('\n         </span>')[1]
    aux = aux.split('<span class="indicator-value">\n          ')[-1]
    base['p/vp'].append(aux) 
    count = count + 1
    if count > 4:
        break
    
```

```python
base
```

```python
df = pd.DataFrame(base)
df['nome_fii'] = nome_fii[:5]
df = df[[ 'nome_fii','liq diaria', 'ultimo rend', 'yield', 'patrim liquido', 'valor patrim',
       'rent mes', 'p/vp']]
df
      
```
